```python
import pandas as pd
import numpy as np
import re
```

```python
cols = {'구분' : '총전입'}
values = '총전출,구내전입,구내전출,구간전입,구간전출,시/도간전입,시/도간전출'.split(',')
for i in range(1,8):
    cols[f'구분.{i}'] = values[i-1]
cols
```

```python
def cleaning_df(year):
    df = pd.read_csv(f'data/data_org/seoul_in_out_{year}_org.csv')
    df = df.drop(0).drop(1)
    print(year)
    print(df.shape)
    print(df.columns)
    df = df.rename(columns=cols)
#     정리된거 저장도 같이 해두자
    df = df.apply(lambda x: x.str.replace(',', ''))
    print(df.iloc[:3,:])
    df.to_csv(f'data/seoul_in_out_{year}.csv', index=False)
    
    return df
```

```python
def concat_dfs(year, result):
    df = cleaning_df(year)
    result = result.append(df, ignore_index=True)
    return result
```

```python
result = pd.DataFrame(columns=['기간', '자치구', '동', '총전입', '총전출', '구내전입',\
                               '구내전출', '구간전입', '구간전출','시/도간전입', '시/도간전출'])
result.head()
```

```python
for year in range(2009, 2018):
    result = concat_dfs(year, result)
result.shape
```

# 여기서부터 돌리기

```python
import pandas as pd
import numpy as np
import re
```

```python
result = pd.DataFrame(columns=['기간', '자치구', '동', '총전입', '총전출', '구내전입',\
                               '구내전출', '구간전입', '구간전출','시/도간전입', '시/도간전출'])
result.head()
```

```python
for year in range(2009, 2018):
    df = pd.read_csv(f'data/seoul_in_out_{year}.csv')
    result = result.append(df, ignore_index=True)
result.head()
```

```python
result.shape
```

```python
result.head()
```

```python
result.to_csv('seoul_in_out_total_abs.csv', index=False)
```

```python
diff_values = result.copy(deep=True)
diff_values = diff_values.iloc[:, 3:]
diff_values.head()
```

```python
new_cols = '총전입-전출,구내전입-전출,구간전입-전출,시/도간전입-전출'.split(',')
for i in range(len(new_cols)):
    diff_values[new_cols[i]] = diff_values.iloc[:,i*2].values - diff_values.iloc[:,i*2+1].values
diff_values.head()
```

```python
diff_values[['기간', '자치구', '동']] = result.iloc[:,:3].copy(deep=True)
diff_values.head()
```

```python
diff_values = diff_values[['기간', '자치구', '동', '총전입', '총전출', '구내전입', '구내전출', '구간전입', '구간전출',\
                  '시/도간전입', '시/도간전출', '총전입-전출', '구내전입-전출', '구간전입-전출', '시/도간전입-전출']]
diff_values.head()
```
